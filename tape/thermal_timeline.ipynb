# 温度の時間変化をプロット

## ポット内温度の定常性確認

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob as glob

filePath = '../../logger_data/check_stationarity/GL100_ab4671_2023-12-27_22-02-36.csv'

df = pd.read_csv(filePath
									#,sep="\t"
									#,encoding="Shift-JIS"
                 ,skiprows=21
                 )
df['Time'] = pd.to_datetime(df['Time'])

df_thermal = df.iloc[:,[1,4,5,6]]
df_thermal.columns = ['Time','90dC','80dC','70dC']
df_thermal.plot(#x='Time',
								y=['90dC', '80dC', '70dC'], kind='line')
plt.xlabel('Time[min]')
plt.ylabel('Temperature')
plt.title('Temperature vs Time')
plt.legend()
plt.grid(True)
plt.show()

ax = plt.axes()
df_thermal.plot(
    title='temp stationarity',
    grid=True,
    legend=True,
    alpha=0.5,
    kind='hist',
    bins=100,
    ax=ax  # Specify the axis for overlaying histograms
)
print("\n ****** mean ******\n",df_thermal[['T1', 'T2', 'T3']].mean())
print("\n ****** stdev ******\n",df_thermal[['T1', 'T2', 'T3']].std(ddof=1))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob as glob


filePath = '../../logger_data/check_peak/*.csv'
#filePath = '../../logger_data/check_peak/GL100_ab4671_2023-12-19_16-17-04.csv'

csv_files = sorted(glob.glob(f'{filePath}'))
for file in csv_files:

	df = pd.read_csv(file
										#,sep="\t"
										,encoding="Shift-JIS"
	                 ,skiprows=21
	                 )
	df['Time'] = pd.to_datetime(df['Time'])
	df_thermal = df.iloc[:,[1,4,5,6]]
	df_thermal.columns = ['Time','Air','SUS','Tape']
	#****** add subtraction ******
	df_thermal['Tape-SUS'] = df_thermal['Tape']-df_thermal['SUS']
	df_thermal['Tape-Air'] = df_thermal['Tape']-df_thermal['Air']
	df_thermal['SUS-Air'] = df_thermal['SUS']-df_thermal['Air']
	print(df_thermal)
	#****** plot ******
	fig ,ax = plt.subplots()
	substraction = 'Tape-SUS'
	df_thermal.plot(y=['Air','SUS','Tape'], kind='line',ax=ax)
	df_thermal.plot(y=[substraction], kind='line',ax=ax)
	plt.xlabel('Time[sec]')
	plt.ylabel('Temperature')
	plt.title(f'DTA:{file}')
	plt.legend()
	#plt.xlim([0,3E3])
	plt.grid(True)
	plt.show()
```
